### Creating Relationships and Merging DataFrames

#### Merging AdventureWorks_Sales with AdventureWorks_Territories and AdventureWorks_Returns

```python
# Merge Sales with Territories
df_sales_territories = pd.merge(df_sales, df_territories, left_on='TerritoryKey', right_on='SalesTerritoryKey', how='left')

# Merge df_sales_territories with df_returns
df_sales_returns = pd.merge(df_sales_territories, df_returns, on=['ProductKey', 'TerritoryKey'], how='left')

# Display the top 100 rows of the merged DataFrame
df_sales_returns
```

```python
# Display the top 20 rows
df_sales_returns.head(20)
```

#### Merging AdventureWorks_Products with AdventureWorks_Product_Categories and AdventureWorks_Product_Subcategories

##### Merge df_products with df_subcategories on ProductSubcategoryKey

##### Merge the result with df_categories on ProductCategoryKey

```python
# Step 1: Merge df_products with df_subcategories
df_products_subcategories = pd.merge(df_products, df_subcategories, left_on='ProductSubcategoryKey', right_on='ProductSubcategoryKey', how='left')

# Step 2: Merge the result with df_categories
df_products_full = pd.merge(df_products_subcategories, df_categories, left_on='ProductCategoryKey', right_on='ProductCategoryKey', how='left')

# Display the merged DataFrame
df_products_full.head()
```

```python
# Set pandas to display all rows
pd.set_option('display.max_rows', None)

# Merge df_products with df_subcategories
df_products_subcategories = pd.merge(df_products, df_subcategories, left_on='ProductSubcategoryKey', right_on='ProductSubcategoryKey', how='left')

# Merge the result with df_categories
df_products_full = pd.merge(df_products_subcategories, df_categories, left_on='ProductCategoryKey', right_on='ProductCategoryKey', how='left')

# Display the full DataFrame
df_products_full
```

#### Merge Sales with Customers

```python
df_sales_customers = pd.merge(df_sales_returns, df_customers, left_on='CustomerKey', right_on='CustomerKey', how='left')
```

### Checking for Null Values and Dropping Unnecessary Columns
